```python
import requests
from bs4 import BeautifulSoup
import shutil
import os
import pandas as pd
```

```python
PATH = './data/'
headers = {'User-Agent': 'Mozilla/5.0'}
df = pd.DataFrame({'filename': [], 'description': [], 'author': []})
url_author = {'Levitan': 'https://gallerix.ru/album/Levitan',
              'Repin': 'https://gallerix.ru/album/Repin',
              'Vasnecov': 'https://gallerix.ru/album/Vasnecov',
              'Serov': 'https://gallerix.ru/album/Serov',
              'Surikov': 'https://gallerix.ru/album/Surikov',
              'Petrov-Vodkin': 'https://gallerix.ru/album/Petrov-Vodkin'}
```

```python
os.mkdir(PATH)
```

```python
def parse_author(url, name):
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text, 'html.parser')
    all_pics = soup.findAll('div', class_='pic')
    
    # os.mkdir(PATH + name)
    for pic in all_pics:
        try:
            url = 'https://gallerix.ru/' + pic.find('a', class_='animsition-link')['href']
            desc_auth = pic.find('div', class_='a_dv1').find('img')['alt']
            description = ' '.join(pic.find('div', class_='a_dv1').find('img')['alt'].split(', ')[:-1])
            author = pic.find('div', class_='a_dv1').find('img')['alt'].split(', ')[-1]

            r = requests.get(url, headers=headers)
            soup = BeautifulSoup(r.text, 'html.parser')
            img_link = soup.find('div', class_='tab-pane animation-fade active').find('a').find('img')['src']
            img_name = img_link.split('/')[-1]

            print('img_url: ',  img_link, 'desc: ', description, 'author: ', author)

            # r = requests.get(img_link, stream=True, headers=headers)
            # with open(PATH + img_name, 'wb+') as f:
            #     r.raw.decode_content = True
            #     shutil.copyfileobj(r.raw, f)
            df.loc[len(df.index)] = [PATH + img_name, description, author] 
        except:
            print('SOS SOS SOS SOS SOS SOS')
```

```python
for author, url in url_author.items():
    print()
    print(author)
    print()
    parse_author(url, author)
```
